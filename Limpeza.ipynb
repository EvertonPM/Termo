```python
import requests
from bs4 import BeautifulSoup

# Função para extrair palavras de uma página
def extrair_palavras(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Imprimir o conteúdo da página para inspecionar
    print(soup.prettify())  # Verifique a estrutura HTML
    
    # Localiza a área que contém as palavras - ajuste de acordo com a estrutura do site
    palavras_html = soup.find_all('a')  # Verifique se as palavras estão dentro de tags <a>
    palavras = [p.text.strip() for p in palavras_html]
    
    return palavras

# Exemplo de como iterar pelas páginas (substitua pelos links corretos das páginas)
urls = [f"https://pt.wikwik.org/palavras5letraspagina{i}.htm" for i in range(2, 16)]

# Armazena todas as palavras
todas_palavras = []

for url in urls:
    palavras = extrair_palavras(url)
    todas_palavras.extend(palavras)
```

```python
import unicodedata

# Sua lista de palavras
palavras = ['ABAJU', 'ABAJÁ', 'AMAPÁ', 'AMBAS', 'AJAJÁ', 'ATUAM', 'ABABÁ', 'AGUAÍ', 'ASSAM', 'ASSAS', 'ASPAS', 'ASSAZ', 'ATUAS']

# Função para remover acentos e manter a letra
def remover_acentos(palavra):
    # Normaliza a string para decompor os caracteres acentuados
    nfkd = unicodedata.normalize('NFKD', palavra)
    # Filtra os caracteres sem acentos
    palavra_sem_acento = ''.join(c for c in nfkd if not unicodedata.combining(c))
    return palavra_sem_acento

# Aplicar a função a cada palavra na lista
palavras_sem_acentos = [remover_acentos(p) for p in palavras_unicas]

# Exibir o resultado
palavras_sem_acentos
```

```python

# Letras proibidas
letras_proibidas = set('REINOCLD')

# Função para filtrar as palavras
def filtrar_palavras(palavras):
    palavras_filtradas = []
    
    for palavra in palavras:
        # Verifica se a palavra tem 5 letras
        if len(palavra) != 5:
            continue
        
        # Verifica se contém duas letras A
        if palavra.count('A') != 2:
            continue
        
        # Verifica se contém letras proibidas
        if any(letra in palavra.upper() for letra in letras_proibidas):
            continue
        
        # Verifica se 'A' não está na segunda e quinta posição
        if palavra[1] == 'A' or palavra[4] == 'A':
            continue
        
        # Se atender a todos os critérios, adiciona à lista
        palavras_filtradas.append(palavra)
    
    return palavras_filtradas

# Aplicar o filtro
resultado = filtrar_palavras(palavras_sem_acentos)

# Exibir o resultado
print(resultado)
```

```python
import csv

# Nome do arquivo CSV
nome_arquivo = 'palavras.csv'

# Salvar a lista em um arquivo CSV
with open(nome_arquivo, mode='w', newline='', encoding='utf-8') as arquivo_csv:
    escritor = csv.writer(arquivo_csv)
    # Escrever cada palavra como uma nova linha no arquivo CSV
    for palavra in palavras_sem_acentos:
        escritor.writerow([palavra])  # Coloca a palavra em uma lista para escrever como uma linha

print(f"As palavras foram salvas em '{nome_arquivo}'.")
```
